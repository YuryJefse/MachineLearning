```r
library("MASS")
```

#### Ones 2x3

```r
matrixOnlyOnes = matrix(sample(1, size = 2*3, replace = TRUE), 2, 3)
matrixOnlyOnes
```

#### Zeros 2x3

```r
matrixOnlyZeros = matrix(sample(0, size = 2*3, replace = TRUE), 2, 3)
matrixOnlyZeros
```

#### Identity 3x3

```r
matrixIdentity = diag(3)
matrixIdentity
```

#### Random Normal Dist. Matrix 2x3

```r
matrixRandNormalDistribution = matrix(rnorm(2*3), 2, 3) 
matrixRandNormalDistribution
```

#### Random Uniform Matrix 2x3

```r
matrixRand = matrix(runif(2*3), 2, 3) 
matrixRand
```

#### Histogram

```r
hist(matrixRandNormalDistribution)
```
